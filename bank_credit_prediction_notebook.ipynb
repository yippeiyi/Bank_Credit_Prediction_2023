### 2. Seperating Data for Training & Testing

Machine learning (ML) is categorised as part of the artificial intelligence (AI) and computer science field as it
focuses primarily on utilizing data and algorithms to allow computers to imitate the learning of humans, while
gradually enhancing precision over time as real-world information is continuously provided. From this, ML
can be further branched into supervised ML, which uses labelled data as input to train the algorithm for
outcome prediction or classification. The term “labelled data” refers to data that is already tagged with
accurate information or labels and can be used as reference for learning by algorithm. The labelled data can
split into training and test datasets in the ratio of ~70% to ~30%. The portion of labelled data that is used as
training data is mainly used by the ML algorithm model for learning. Whereas the portion of labelled data
that is used as testing data is mainly used to evaluate the accuracy and performance of the trained ML
algorithm.

The label for the data, which will be identified as y variable, is the credit score. Whereas the features for the
data, which will be identified as x variable. 

```python
exclude = ['ID','Month','Age','Occupation','Monthly_Inhand_Salary','Amount_invested_monthly','Credit_Score','Annual_Income']

# Assign features and y
x = dataset.drop(exclude, axis=1)
y = dataset["Credit_Score"]

# Split data into training and testing set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x.values,y.values, test_size=0.3, random_state=0)
```

### 3. Feature Scaling/Normalization

Regarding supervised ML, it can further branch to classification. Classification uses algorithm that assign
output into specific categories or classes. When algorithm groups output into 2 distinct classes, this is known
as binary classification. However, if algorithm groups output into more than 2 distinct classes, this is known
as multi-class classification. Some common examples of classification algorithms are decision trees, random
Forests, support vector machines (SVM), and logistic regression.

Prior to building the SVM model, it is essential that dataset is normalised/scaled. This is because initial
dataset range varies widely, which therefore affects the ability of some objective functions to not work
properly. For example, many classifiers use the Euclidean distance which is the distance between 2 points. If
one of the features were to have a wide range of values, distance would be heavily influenced by the feature.
Therefore, by scaling/normalizing the data into a common scale, it reduces the impact of potential outliers
and therefore improve accuracy and stability of ML model.

```python
# Scale the dataset
from sklearn.preprocessing import StandardScaler
sc = StandardScaler().fit(x_train)
x_train = sc.transform(x_train)
x_test = sc.transform(x_test)
```

###  4. Building the SVM Model 

To distinguish customers into a multi-class of specific credit scores (1-Poor, 2-Standard, 3-Good), the SVM
algorithm is used. SVM algorithm, which was developed by Vladimir Vapnik, can be used for either regression
or (most of the time) classification problems. The data points in SVM algorithm are plotted in n-dimensional
space (where n is the no. of features) and is represented as a coordinate. The term “support vectors” simply
refers to coordinates of each individual observation that are closest to hyperplane.

A Hyperplane can be described as the decision boundary which separates all data points into specific classes.
Construction of optimal hyperplanes are based on the maximum distance between 2 classes of data. The
term “margin” refers to the distance between nearest data points and hyperplane. The algorithm is named
as SVM because it chooses support vectors to create a hyperplane. SVM classification can be considered as
robust against outliers, this is because the algorithm has a feature to ignore outliers and find hyperplane with
the maximum margin.

In SVM algorithm, there is a technique known as the SVM kernel trick which aims to use a linear classifier to
solve an existing non-linear problem by implementing the use of a group of mathematical functions (e.g.,
linear, polynomial, gaussian radial basis function, sigmoid) which takes low dimensional input space and
transforms it into a higher dimensional input space. In other words, after the transformation to higher
dimensional space, the classes can be linearly separated in this higher dimension space as compared to its
previous low dimensional space. 


```python
# Model Building & Training
from sklearn.svm import SVC
svc = SVC(kernel='rbf', C=1, gamma=0.1, decision_function_shape='ovr')
svc = svc.fit(x_train,y_train)
y_pred = svc.predict(x_test)
```

### 5. Model Accuracy (Confusion Matrix & QWK)

After the training and testing of model, the accuracy was analysed in the form of a 3x3 confusion matrix and Quadratic Weighted Kappa (QWK) score using the `sklearn.metrics` modules.

```python
# Evaluating Model Prediction
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, accuracy_score, cohen_kappa_score
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print('Accuracy is  : %.3f' % accuracy_score(y_test,y_pred))
print("Quadratic Weighted Kappa score: %.3f" %cohen_kappa_score(y_test, y_pred, labels=None, weights="quadratic"))
```

Confusion matrix is computed and shown as figure below:

```python
# Plot the confusion matrix for model:
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

c_matrix = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(c_matrix,
                     index = [1,2,3], 
                     columns = [1,2,3])

#Plotting the confusion matrix
plt.figure(figsize=(5,4))
sns.heatmap(cm_df, annot=True)
plt.title('Confusion Matrix')
plt.ylabel('Predicted')
plt.xlabel('Actual')
plt.show()
```

The True Positive (TP), True Negative (TN), False Positive (FP), and False Negative (FN) values from the 3x3 confusion matrix are calculated shown as a table below:

```python
import numpy as np

# Confusion Matrix Values for Credit Score:1
data = {
    "TP":c_matrix[0,0],
    "TN":[np.sum(c_matrix[1:,1:])],
    "FP":[np.sum(c_matrix[0,1:])],
    "FN":[np.sum(c_matrix[1:,0])]
}

credit_score = pd.DataFrame(data)

credit_score.insert(loc = 0,
                   column = "Credit Score",
                   value = 1)

##############################################
# Confusion Matrix Values for Credit Score:2
data2 = {
    "TP":c_matrix[1,1],
    "TN":[np.sum(c_matrix[0::2,0::2])],
    "FP":[np.sum(c_matrix[1,0::2])],
    "FN":[np.sum(c_matrix[0::2,1])]
}

credit_score_2 = pd.DataFrame(data2)

credit_score_2.insert(loc = 0,
                      column = "Credit Score",
                      value = 2)

##############################################
# Confusion Matrix Values for Credit Score:3
data3 = {
    "TP":c_matrix[2,2],
    "TN":[np.sum(c_matrix[:2,:2])],
    "FP":[np.sum(c_matrix[2,:2])],
    "FN":[np.sum(c_matrix[:2,2])]
}

credit_score_3 = pd.DataFrame(data3)

credit_score_3.insert(loc = 0,
                      column = "Credit Score",
                      value = 3)

##############################################
# Concatenate the 3 tables for credit score:
new_matrix = pd.concat([credit_score.reset_index(drop=True),credit_score_2.reset_index(drop=True),credit_score_3.reset_index(drop=True)], ignore_index=True)
new_matrix
```

Explanation of TP, TN, FP, FN:
> 1. TP: Correctly predicts that credit score is 1, 2, or 3.
> 2. TN: Correctly predicts that credit score is not 1, 2, or 3.
> 3. FP: Incorrectly predicts that credit score is 1, 2, or 3.
> 4. FN: Incorrectly predicts that credit score is not 1, 2, or 3.


The quadratic weighted kappa (QWK) is a metric that measures the agreement between 2 outcomes. To further elaborate, it is the amount of agreement between an algorithm’s prediction and true labels. The range of QWK score is from 0.0 (random agreement) to 1.0 (perfect agreement). There are also situations where QWK score can go below 0.0 due to less agreement expected by chance.
